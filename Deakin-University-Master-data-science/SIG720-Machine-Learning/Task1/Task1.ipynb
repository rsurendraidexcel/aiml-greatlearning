```python
#Normalize Data Data
from sklearn.preprocessing import normalize
data_scaled = normalize(df)
df = pd.DataFrame(data_scaled, columns=df.columns)
df.head()
```

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler
scale = StandardScaler()
X_scale = scale.fit_transform(df)
df = pd.DataFrame(X_scale, columns=df.columns)
df.head()
```

```python
# Generate synthetic dataset with 64 column
X=df.iloc[:,0:-1]
y=df.iloc[:,-1]
```

```python
X.shape
```

```python
sns.scatterplot(data=X);
```

```python
#Finding Silhoetee Score and K values
from sklearn.metrics import silhouette_score
range_n_clusters =  [x+2 for x in range(64)]
silhouette_avg = []
for num_clusters in range_n_clusters:
    # initialise kmeans
    skmeans= KMeans(n_clusters=num_clusters, n_init='auto', init = 'k-means++', random_state = 42)
    skmeans.fit(X)
    cluster_labels = skmeans.labels_
    #silhouette score
    silhouette_avg.append(silhouette_score(X, cluster_labels))

plt.plot(range_n_clusters, silhouette_avg,'bx-')
plt.xlabel('Values of K') 
plt.ylabel('Silhouette score') 
plt.title('Silhouette analysis For Optimal k')
plt.show()
```

```python
#K-means clustering 50 time repeting
inertias = []
for i in range(1,50):
    kmeans = KMeans(n_clusters=i, n_init='auto', init='k-means++'  )
    kmeans.fit(data)
    inertias.append(kmeans.inertia_)
```

```python
#Printing the Elbow Method chart  Finding the optimal K value 
plt.plot(range(1,50), inertias, marker='o')
plt.title('Elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('No of Features')
plt.show()
```

```python
## loding "diabetes" datasets from SKlearn
from sklearn.datasets import load_diabetes
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
DX,Dy =load_diabetes(return_X_y = True , as_frame = True)
```

```python
# diabetes = load_diabetes()
# df = pd.DataFrame(data=diabetes.data, columns=diabetes.feature_names)
```

```python
DX
```

```python
Dy
```

```python
DX.shape, Dy.shape
```

```python
#PCA Model Define
scaler = StandardScaler()
scaler.fit(DX)
DX_scaled = scaler.transform(DX)
pca = PCA(n_components=3)
pca.fit(DX_scaled)
```

```python
eigenvalues = pca.explained_variance_
eigenvectors = pca.components_
```

```python
print("eigenvalues", eigenvalues)
```

```python
print("eigenvectors", eigenvectors)
```

```python
for v in pca.explained_variance_ratio_:
    print('Explained variation per principal component: {}%'.format(round(v*100,2)))
```

```python
pcadf=pd.DataFrame(pca.components_,columns=list(DX.columns))
pcadf.transpose()
```

```python
# Principal components correlation coefficients
pca.components_
ax = sns.heatmap(DX.corr(), annot=True)
```

```python
PCA_Scores = pd.DataFrame(pca.fit_transform(DX_scaled), columns = ['PC 1', 'PC 2', 'PC 3'])
PCA_Scores.head()
```

```python
#Determine explained variance
total_eigenvalues = sum(eigenvalues)
var_exp = [(i/total_eigenvalues) for i in sorted(eigenvalues, reverse=True)]
```

```python
var_exp
```

```python
total_eigenvalues
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
cum_sum_exp = np.cumsum(var_exp, dtype=float)
plt.bar(range(0,len(var_exp)), var_exp, alpha=0.5, align='center', label='Individual explained variance')
plt.step(range(0,len(cum_sum_exp)), cum_sum_exp, where='mid',label='Cumulative explained variance')
plt.ylabel('Explained variance ratio')
plt.xlabel('Principal component index')
plt.legend(loc='best')
plt.tight_layout()
plt.show()
```

```python
#Row data Plating in 3d space
ax = plt.figure().add_subplot(projection='3d')
ax.scatter(DX['age'], DX['sex'], DX['bmi'], c=Dy)
ax.scatter(DX['s1'], DX['s2'], DX['s3'], c=Dy )
ax.scatter(DX['s4'], DX['s5'], DX['s6'], c=Dy)
plt.show()
```

```python
colors=['b', 'r', 'g'] 
fig = plt.figure(1)
ax = fig.add_subplot(111, projection='3d')
# Let's set up the first dataset of actives_df. 
p1 = ax.plot(DX['sex'], 
             DX['age'], 
             DX['bmi'], 
             'o', color=colors[0],                                   
             alpha = 0.6, label='Feature 1',  
             markersize=3, 
             markeredgecolor='black',
             markeredgewidth=0.1)

# Let's set up the second dataset of micro_df.
p2 = ax.plot(DX['bp'], 
             DX['s1'], 
             DX['s2'], 
             '*', color=colors[1],                           
             alpha = 0.6, label='Feature 2',
             markersize=3, 
             markeredgecolor='black',
             markeredgewidth=0.1)
# Let's set up the third datset of approved_df.
p3 = ax.plot(DX['s3'], 
             DX['s4'], 
             DX['s5'], 
             'p', color=colors[2],                                     
             alpha = 0.6, label='Feature 3',   
             markersize=3, markeredgecolor='black',
             markeredgewidth=0.1)
# We will then label the three axes using the percentages explained for each major component.
ax.set_xlabel('PCA-1, ' +  str(round(pca.explained_variance_ratio_[0]*100,2)) + '% Explained', fontsize=6)
ax.set_ylabel('PCA-2, ' +  str(round(pca.explained_variance_ratio_[1]*100,2)) + '% Explained', fontsize=6)
ax.set_zlabel('PCA-3, ' +  str(round(pca.explained_variance_ratio_[2]*100,2)) + '% Explained', fontsize=6)

fig.legend(fontsize = 'x-small', loc='upper center', markerscale=2)
plt.autoscale()
plt.rcParams["figure.dpi"] = 2000                                   # set the figure resolution dpi value to 1000
plt.show()
fig_name = '3D_scatterplot_PCA.png'
fig.savefig(fig_name)
```
